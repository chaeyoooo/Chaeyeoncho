# 국가, 장르 별 전처리

```python
import pandas as pd

# CSV 파일 불러오기
file_path = '/content/데분.csv'
data = pd.read_csv(file_path , encoding='cp949')  # 한글 데이터를 포함하고 있으므로 'cp949' 인코딩 사용

# 필요한 컬럼만 선택
data_filtered = data[['country', 'listed_in']].dropna()

# 결측치 처리
data_filtered['country'].fillna('Unknown', inplace=True)  # 결측치를 'Unknown'으로 대체
data_filtered['listed_in'].fillna('No Category', inplace=True)  # 결측치를 'No Category'로 대체

# 'listed_in' 컬럼에서 카테고리를 쉼표로 구분하여 처리
data_filtered['listed_in'] = data_filtered['listed_in'].str.split(', ')
data_exploded = data_filtered.explode('listed_in')

# 중복 데이터 제거
data_exploded.drop_duplicates(inplace=True)

# 결과 확인
print(data_exploded.head())
print(data_exploded['listed_in'].unique())  # 카테고리 확인

# CSV 파일로 저장할 경우
#data_exploded.to_csv('/path/to/your/output_file.csv', index=False)
```

# 국가별 인기 장르 집계

```python
import pandas as pd

# 파일 경로
file_path = '/content/데분.csv'

# 파일 불러오기 (ISO-8859-1 인코딩 사용)
data = pd.read_csv(file_path, encoding='ISO-8859-1')

# 관련 컬럼 선택 및 결측치 제거
filtered_data = data[['country', 'listed_in']].dropna()

# 국가별로 데이터 분리 (국가명이 여러 개인 경우 분리)
filtered_data['country'] = filtered_data['country'].apply(lambda x: [country.strip() for country in x.split(',')])
filtered_data = filtered_data.explode('country')

# 장르별로 데이터 분리 (여러 장르가 한 셀에 있는 경우 분리)
filtered_data['listed_in'] = filtered_data['listed_in'].apply(lambda x: [genre.strip() for genre in x.split(',')])
filtered_data = filtered_data.explode('listed_in')

# 유효한 국가 이름만 필터링 (영문자 및 공백만 포함된 문자열)
valid_countries = filtered_data['country'].str.contains('^[A-Za-z ]+$')
filtered_valid_countries = filtered_data[valid_countries]

# 각 국가별로 가장 인기 있는 장르 집계
popular_genre_by_country = filtered_valid_countries.groupby('country')['listed_in'].agg(
    lambda x: x.value_counts().idxmax())

# 결과 출력
print(popular_genre_by_country.head(20))
```

#국가별 장르 선호 빈도수 분석

```python
import pandas as pd

# CSV 파일 불러오기
data_path = '/content/데분.csv'
data = pd.read_csv(data_path, encoding='cp949')  # 인코딩을 'cp949'로 설정

# 'country'와 'listed_in' 컬럼에 집중
data_filtered = data[['country', 'listed_in']].dropna()

# 'country' 컬럼 정제하기
data_filtered['country'] = data_filtered['country'].str.split(', ')
data_exploded_countries = data_filtered.explode('country')
data_exploded_countries['country'] = data_exploded_countries['country'].str.strip()
valid_countries = data_exploded_countries['country'].str.contains('^[a-zA-Z ]+$')
cleaned_data = data_exploded_countries[valid_countries]

# 'listed_in' 컬럼 처리하기
cleaned_data['listed_in'] = cleaned_data['listed_in'].str.split(', ')
cleaned_data_exploded = cleaned_data.explode('listed_in')

# 국가별로 'listed_in' 카테고리 빈도수 계산
country_genre_counts = cleaned_data_exploded.groupby(['country', 'listed_in']).size().unstack(fill_value=0)

# 결과 출력
print(country_genre_counts.head())  # 처음 몇 줄만 출력
```

#gdp, 장르, 국가 병합

```python
import pandas as pd

# 파일 경로
content_data_path = '/content/데분.csv'
gdp_data_path = '/content/나라별gdp.csv'

# CSV 파일 불러오기
content_data = pd.read_csv(content_data_path, encoding='cp949')
gdp_data = pd.read_csv(gdp_data_path, encoding='cp949')

# 필요한 컬럼 선택 및 결측치 제거
content_filtered = content_data[['country', 'listed_in']].dropna()
gdp_filtered = gdp_data.rename(columns={'Country': 'country', 'GDP 2022': 'gdp'}).dropna()

# 'listed_in' 컬럼의 데이터를 쉼표로 분할
content_filtered['listed_in'] = content_filtered['listed_in'].str.split(', ')
content_exploded = content_filtered.explode('listed_in')

# 두 데이터셋 병합
merged_data = pd.merge(content_exploded, gdp_filtered, on='country', how='left')

# 결과 출력
print(merged_data.head())
print(merged_data.info())

# 병합된 데이터를 CSV 파일로 저장
#merged_data.to_csv('/path/to/your/merged_output.csv', index=False)
```
